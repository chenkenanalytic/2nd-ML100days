```python
import pandas as pd
from sklearn import datasets, metrics
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv("data/data-science-london-scikit-learn/train.csv", header = None)
train_y = pd.read_csv("data/data-science-london-scikit-learn/trainLabels.csv", header = None)
test = pd.read_csv("data/data-science-london-scikit-learn/test.csv", header = None)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(train, train_y, test_size=0.3, random_state=2)

GBR = GradientBoostingRegressor()

GBR.fit(X_train, Y_train)

y_pred = GBR.predict(X_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(Y_test, y_pred)
print("Score: ", score)
```

```python
y_predictions = GBR.predict(test)
```
